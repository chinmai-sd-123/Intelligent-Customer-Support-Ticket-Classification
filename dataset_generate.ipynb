```python
import pandas as pd


```

```python
labels=[
    "Payment",
    "Refund",
    "Account",
    "Technical",
    "Other"
]
```

```python
import  random
random.seed(42)
```

```python
payment_templates=[
    "payment was deducted but order failed",
    "charged twice during checkout",
    "amount debited but transaction unsuccessful",
    "bank shows debit but no order created",
    "payment completed but app crashed",
    "money taken but order not confirmed",
    "transaction failed after payment",
    "payment successful but order missing",
    "charged for an order that did not complete",
    "checkout failed after amount was deducted"
]
```

```python
refund_templates = [
    "refund not received after cancellation",
    "waiting for refund since last week",
    "refund promised but not credited",
    "refund still pending",
    "no update on refund request",
    "refund process taking too long",
    "money not refunded yet",
    "refund missing after return",
    "refund delayed beyond expected time",
    "refund status not updated"
]
```

```python
account_templates = [
    "account locked after login attempt",
    "cannot access my account",
    "account blocked without reason",
    "login not working anymore",
    "unable to sign in to account",
    "account suspended suddenly",
    "password reset not working",
    "account verification failed",
    "account access denied",
    "unable to log in after update"
]
```

```python
technical_templates = [
    "app crashes on checkout",
    "application freezes frequently",
    "website not loading properly",
    "error occurs while placing order",
    "screen goes blank on payment",
    "app not responding",
    "technical error during checkout",
    "system failure while ordering",
    "page keeps refreshing",
    "unexpected error message appears"
]
```

```python
other_templates = [
    "need help with discount codes",
    "how to update profile information",
    "general question about services",
    "need assistance with offers",
    "query regarding subscription plans",
    "how to change email address",
    "question about loyalty points",
    "help needed with account settings",
    "inquiry about new features",
    "support needed for general issue"
]
```

```python
noise_phrases = [
    "urgent",
    "asap",
    "please help",
    "really frustrated",
    "this is unacceptable",
    "kindly check",
    "need this resolved today",
    "immediately",
    "seriously",
    "not a scam"
]

obfuscation_map = {
    "a": "@",
    "o": "0",
    "i": "1",
    "e": "3",
    "s": "5"
}
```

```python
def add_noise(text):
    if random.random() < 0.5:
        text=text+" "+random.choice(noise_phrases)

    new_text = ""
    for char in text:
        if char in obfuscation_map and random.random() < 0.25:
            new_text += obfuscation_map[char]
        else:
            new_text += char
    text = new_text

    if random.random() < 0.3:
        text = text.upper()
    elif random.random() < 0.3:
        text = text.capitalize()
    return text

```

```python
data=[]

Templates={
    "Payment": payment_templates,
    "Refund": refund_templates,
    "Account": account_templates,
    "Technical": technical_templates,
    "Other": other_templates
}
samples=10 #per template

for category,templates in Templates.items():
    for template_id,base_text in enumerate(templates):
        for i in range(samples):
            noisy_text = add_noise(base_text)
            data.append({
                "ticket_text":noisy_text,
                "category":category,
                "template_id":f"{category}_{template_id}"
            })
df=pd.DataFrame(data)
df=df.sample(frac=1,random_state=42).reset_index(drop=True)
```
